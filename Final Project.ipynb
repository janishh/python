```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path_od = "https://raw.githubusercontent.com/janishh/dataset/main/order_detail.csv"
path_pd = "https://raw.githubusercontent.com/janishh/dataset/main/payment_detail.csv"
path_cd = "https://raw.githubusercontent.com/janishh/dataset/main/customer_detail.csv"
path_sku = "https://raw.githubusercontent.com/janishh/dataset/main/sku_detail.csv"

df_od = pd.read_csv(path_od)
df_pd = pd.read_csv(path_pd)
df_cd = pd.read_csv(path_cd)
df_sku = pd.read_csv(path_sku)
```

```python
df_pd.head(10)
```

```python
from sqlite3 import connect
conn = connect(':memory:')
df_od.to_sql ('order_detail',conn, index='False', if_exists = 'replace')
df_pd.to_sql ('payment_detail',conn, index='False', if_exists = 'replace')
df_cd.to_sql ('customer_detail',conn, index='False', if_exists = 'replace')
df_sku.to_sql ('sku_detail',conn, index='False', if_exists = 'replace')

```

```python
df = pd.read_sql ("""
SELECT
    order_detail.*,
    payment_detail.payment_method,
    sku_detail.sku_name,
    sku_detail.base_price,
    sku_detail.cogs,
    sku_detail.category,
    customer_detail.registered_date
FROM order_detail
LEFT JOIN payment_detail
    on payment_detail.id = order_detail.payment_id
LEFT JOIN sku_detail
    on sku_detail.id = order_detail.sku_id
LEFT JOIN customer_detail
    on customer_detail.id = order_detail.customer_id
""", conn
)
```

```python
df.dtypes
```

```python
df = df.astype ({'before_discount':'int','discount_amount':'int','after_discount':'int', 'base_price':'int'})
df.dtypes
```

```python
df['order_date'] = pd.to_datetime (df['order_date'])
df['registered_date'] = pd.to_datetime (df['registered_date'])
df.dtypes
```

```python
#TOP 5 Produk dari Kategori Mobiles & Tablets selama tahun 2022, dengan jumlah kuantitas penjualan (valid = 1) paling tinggi.

data1 = pd.DataFrame(\
    
                    df[(df['is_valid']==1) &\
                    (df['category']=='Mobiles & Tablets') &\
                    ((df['order_date'] >= '2022-01-01') & (df['order_date'] <= '2022-12-31'))]\
                    .groupby(by=["sku_name"])["qty_ordered"]\
                    .sum()\
                    .head(5)\
                    .sort_values(ascending=False)\
                    
.reset_index(name = 'qty_2022'))
    
```

```python
data1.head()
```

```python
#pengecekan data penjualan kategori tersebut dengan tahun 2021 secara kuantitas penjualan. 
#Dugaan sementara kami, telah terjadi penurunan kuantitas penjualan pada 2022 dibandingkan 2021. (Mohon juga menampilkan data ke-15 kategori)

data2 = pd.DataFrame(\
    df[(df['is_valid'] == 1) &\
    ((df['order_date'] >= '2022-01-01') & (df['order_date'] <= '2022-12-31'))]\
    .groupby(by=['category'])['qty_ordered']\
    .sum()\
    .sort_values(ascending=False)\
    .reset_index(name='qty_2022')
    )
data2

data3 = pd.DataFrame(\
    df[(df['is_valid'] == 1) &\
    ((df['order_date'] >= '2021-01-01') & (df['order_date'] <= '2021-12-31'))]\
    .groupby(by=['category'])['qty_ordered']\
    .sum()\
    .sort_values(ascending=False)\
    .reset_index(name='qty_2021')
    )
data3

data4 = data2.merge(data3, left_on = 'category', right_on='category')
data4
#Growth = tahun terakhir - tahun sebelumnya
data4['qty_growth']=data4['qty_2022'] - data4['qty_2021']
data4.sort_values(by=['qty_growth'],ascending=True,inplace=True)
data4
```

```python
#Apabila memang terjadi penurunan kuantitas penjualan pada kategori Others, 
#kami mohon bantuan untuk menyediakan data TOP 20 nama produk yang mengalami penurunan paling tinggi pada 2022 jika dibanding dengan 2021. 
#Hal ini kami gunakan sebagai bahan diskusi pada meeting selanjutnya.

ata5 = pd.DataFrame(\
                     #Memfilter data yang sudah melakukan pembayaran (is_valid =1)
                     df[(df['is_valid']==1) &\
                      #Filter Category "Others" karena mengalami penurunan
                      (df['category']=='Others') &\
                    #Memfilter data pada transaksi selama 2021
((df['order_date'] >= '2021-01-01') & (df['order_date'] <= '2021-12-31'))]\
                     #Mengelompokan data
                     .groupby(by=['sku_name'])["qty_ordered"]\
                     #Pengelompokan berdasarkan penjumlahan
                     .sum()\
                      #Mengurutkan Data
                     .sort_values(ascending=False)\
                     #Reset nama header
                     .reset_index(name='qty_2021'))
data5
```
